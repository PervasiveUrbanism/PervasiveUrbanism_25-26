## Simple Graph

```python
import seaborn as sns
import pandas as pd



gsrPath = "010RC14X.CSV"

df = pd.read_csv(filepath_or_buffer = gsrPath, sep='\t')

df

```

```python
listCount = df["Count"].tolist()

listGSR = df["GSR"].tolist()

# sns.set_theme(style="darkgrid")
sns.set_theme(style="whitegrid")


lineplot = sns.lineplot(x=listCount, y=listGSR, data=df)

lineplot.figure.savefig("lineplot.pdf")
```

## Bar Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = pd.read_csv("Excercise 03 - Flickr.csv")
df.head(3)
```

Each image can possess numerous tags, located within the "tags" column, separated by a space. To be interpreted by Python, it needs to be converted into a list format, as shown below:

`signs london olympus lee c740 wysiwyg`  &#8594;  `["signs","london","olympus","lee", "c740","wysiwyg"]`

And this is what the `.split()` command does.

```python
# column_values = df["tags"]
column_values = df["tags"].str.split()

# column_values.head(3)

flat = column_values.explode()
 
counts=flat.value_counts()

dateset=counts.iloc[1:40]
ax = dateset.plot(kind='bar');
#this row saves it as a pdf:
ax.figure.savefig("graph.pdf")
```

```python
# look at the count of accuracy

listAccuracy = df["accuracy"].tolist()

sns.histplot(listAccuracy, kde=True)
plt.show()
```
